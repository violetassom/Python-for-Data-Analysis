### "ÏÑúÏö∏ÏãúÏóêÎäî Íµ¨Î≥ÑÎ°ú ÌèâÍ∑† ÏïΩ 400 Í∑∏Î£® Ï†ïÎèÑÏùò `Ïù¥ÌåùÎÇòÎ¨¥`Í∞Ä ÏûàÎã§"ÎùºÎäî Í∞ÄÏÑ§Ïóê ÎåÄÌï¥ Í∞ÄÏÑ§Í≤ÄÏ†ïÏùÑ ÏãúÌñâÌïòÏÑ∏Ïöî.

- Îã®, `ÏÑúÏö∏Ïãú`, `Í≥µÏõêÎÖπÏßÄÏÇ¨ÏóÖÏÜå`, `ÏãúÏÑ§Í¥ÄÎ¶¨Í≥µÎã®` Î∂ÄÎ∂ÑÏùò Îç∞Ïù¥ÌÑ∞Îäî Ï†úÏô∏Ìï©ÎãàÎã§.
- Í∞ÄÏÑ§Í≤ÄÏ†ï Í≤∞Í≥ºÏùò pvalue Í∞íÏùÑ `pv1`Ïóê Ï†ÄÏû•ÌïòÏÑ∏Ïöî. (`.pvalue` ÏÇ¨Ïö©)

```python
new_header = seoul.iloc[0]
seoul = seoul[1:]
seoul.columns = new_header
seoul.head()
```

```python
cond = ((seoul['ÏûêÏπòÍµ¨']=='ÏÑúÏö∏Ïãú')|(seoul['ÏûêÏπòÍµ¨']=='Í≥µÏõêÎÖπÏßÄÏÇ¨ÏóÖÏÜå')|(seoul['ÏûêÏπòÍµ¨']== 'ÏãúÏÑ§Í¥ÄÎ¶¨Í≥µÎã®'))
print(seoul.shape)
seoul = seoul[~cond]
seoul.shape
```

```python
def remove_com(string):
  return string.replace(',','')
```

```python
def conver_int(string):
  return int(string)
```

```python
def convert_zero(missing):
  return missing.replace('-','0')
```

```python
seoul = seoul.reset_index()
seoul = seoul.iloc[:,1:]
seoul = seoul.fillna(0)
seoul.head()
```

```python
ran = seoul.shape[1]

for i in range(2,ran):
  seoul.iloc[:,i] = seoul.iloc[:,i].apply(remove_com)
  seoul.iloc[:,i] = seoul.iloc[:,i].apply(convert_zero)
  seoul.iloc[:,i] = seoul.iloc[:,i].apply(conver_int)
```

```python
seoul.dtypes
```

```python
from scipy import stats
```

```python
# Ïù¥ Í≥≥Ïóê ÎãµÏïàÏùÑ ÏûëÏÑ±ÌïòÏãúÍ∏∞ Î∞îÎûçÎãàÎã§
pv1 = stats.ttest_1samp(seoul['Ïù¥ÌåùÎÇòÎ¨¥'],400).pvalue
pv1
```


### Ïù¥Î≤àÏóêÎäî 10Í∞úÏùò Íµ¨Î•º ÏÉòÌîåÎßÅÌïòÏó¨, ÏúÑÏóê ÎåÄÌïú Í∞ÄÏÑ§Í≤ÄÏ†ïÏùÑ Îã§Ïãú ÌïúÎ≤à ÏãúÌñâÌï¥Î≥¥ÏÑ∏Ïöî.

- Îã®, ÏïÑÎûòÏùò ÏΩîÎìúÎ•º Ï∞∏Ï°∞ÌïòÏó¨ sampleÏùÑ ÏúÑÌïú seed Îäî 123Î°ú ÏÑ§Ï†ïÌïòÏÑ∏Ïöî.
- Í∞ÄÏÑ§Í≤ÄÏ†ï Í≤∞Í≥ºÏùò pvalueÍ∞íÏùÑ `pv2`Ïóê Ï†ÄÏû•ÌïòÏÑ∏Ïöî


```python
import numpy as np
np.random.seed(123)

cond = np.random.choice(seoul['ÏûêÏπòÍµ¨'], size = 10) # "~"ÏôÄ " "ÏùÑ Ï†ÅÏ†àÌïú Í∞íÏúºÎ°ú Ï±ÑÏö∞ÏãúÎ©¥ Îê©ÎãàÎã§
cond
```

```python
seoul_sample = seoul[seoul['ÏûêÏπòÍµ¨'].isin(cond)]
seoul_sample
```

```python
pv2 = stats.ttest_1samp(seoul_sample['Ïù¥ÌåùÎÇòÎ¨¥'],400).pvalue
pv2
```
