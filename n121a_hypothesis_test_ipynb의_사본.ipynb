### "서울시에는 구별로 평균 약 400 그루 정도의 `이팝나무`가 있다"라는 가설에 대해 가설검정을 시행하세요.

- 단, `서울시`, `공원녹지사업소`, `시설관리공단` 부분의 데이터는 제외합니다.
- 가설검정 결과의 pvalue 값을 `pv1`에 저장하세요. (`.pvalue` 사용)

```python
new_header = seoul.iloc[0]
seoul = seoul[1:]
seoul.columns = new_header
seoul.head()
```

```python
cond = ((seoul['자치구']=='서울시')|(seoul['자치구']=='공원녹지사업소')|(seoul['자치구']== '시설관리공단'))
print(seoul.shape)
seoul = seoul[~cond]
seoul.shape
```

```python
def remove_com(string):
  return string.replace(',','')
```

```python
def conver_int(string):
  return int(string)
```

```python
def convert_zero(missing):
  return missing.replace('-','0')
```

```python
seoul = seoul.reset_index()
seoul = seoul.iloc[:,1:]
seoul = seoul.fillna(0)
seoul.head()
```

```python
ran = seoul.shape[1]

for i in range(2,ran):
  seoul.iloc[:,i] = seoul.iloc[:,i].apply(remove_com)
  seoul.iloc[:,i] = seoul.iloc[:,i].apply(convert_zero)
  seoul.iloc[:,i] = seoul.iloc[:,i].apply(conver_int)
```

```python
seoul.dtypes
```

```python
from scipy import stats
```

```python
# 이 곳에 답안을 작성하시기 바랍니다
pv1 = stats.ttest_1samp(seoul['이팝나무'],400).pvalue
pv1
```


### 이번에는 10개의 구를 샘플링하여, 위에 대한 가설검정을 다시 한번 시행해보세요.

- 단, 아래의 코드를 참조하여 sample을 위한 seed 는 123로 설정하세요.
- 가설검정 결과의 pvalue값을 `pv2`에 저장하세요


```python
import numpy as np
np.random.seed(123)

cond = np.random.choice(seoul['자치구'], size = 10) # "~"와 " "을 적절한 값으로 채우시면 됩니다
cond
```

```python
seoul_sample = seoul[seoul['자치구'].isin(cond)]
seoul_sample
```

```python
pv2 = stats.ttest_1samp(seoul_sample['이팝나무'],400).pvalue
pv2
```

## 2. Two-Sample T-test 연습

### "서울시의 구별 평균 느티나무수와 왕벚나무의 수는 차이가 없다"라는 가설에 대해 가설검정을 시행하세요.

- 단, `서울시`, `공원녹지사업소`, `시설관리공단` 부분의 데이터는 제외합니다.
- 가설검정 결과의 pvalue를 `pv3`에 저장하세요

```python
# 이 곳에 답안을 작성하시기 바랍니다
pv3 = stats.ttest_ind(seoul['느티나무'],seoul['왕벚나무']).pvalue
pv3
```


### Two-Sample One-tailed test를 사용하여 **평균 느티나무수가 왕벚나무 수보다 크다**라는 가설에 대한 가설검정을 시행하세요.
- 해당 가설검정의 결과를 `ans`에 저장하세요.

이 때 `ans`가 가질 수 있는 값은 아래와 같습니다 :
1. "느티나무" ( 느티나무의 수가 통계적으로 유의하게 많은 경우 )
2. "왕벚나무" ( 왕벚나무의 수가 통계적으로 유의하게 많은 경우 )
3. "없음" (통계적으로 유의하지 않은 경우)

Two-Sample One-tailed test를 scipy에서 하는 방법은 아래의 링크를 참고 하세요  
참고 링크 : https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html

```python
stats.ttest_ind(seoul['느티나무'],seoul['왕벚나무'],alternative='greater').pvalue
```

```python
# 이 곳에 답안을 작성하시기 바랍니다
ans = '느티나무'
ans
```
